```python
#Task one of Project 1: Approximate solution to differential equation of form dy(t)/dt = -kt

import numpy as np
import matplotlib.pyplot as plt

T = 10      #Value at which to end simulation
dt = 0.5    #Smaller dt leads to smoother curve and closer approximation to known function result
k = 1   #Constant, smaller = shallower decay
y0 = 100     #Starting value of y
t = [0]     #Time variable array with start value 0 
y = [y0]    #Y variable array with start value y0
yex = [y0]  #Initial time for exp function

nt = int(T/dt) #number of steps between start and end value

for i in range(1,nt): #for index in range between 1 and the number of steps in total
    t.append(i*dt)    #add i.dt per step in t direction t the time array
    y.append(y[i-1]-dt * k * y[i-1]) #add this in the y array
    yex.append(y0*np.exp(-k*i*dt)) #known result for comparison of approximation

plt.rc('font', family = 'serif', serif = 'cmr10',size=16) #to make like latex
plt.xlabel("Time (s)") #add x axis label
plt.ylabel("Position") #add y axis label
plt.legend(loc='upper right') #adds legend
plt.plot(t,y,"yo--", label = "Euler") #plot approx function as blue crosses
plt.plot(t,yex,"bx", label = "Analysis") #plot known solution as red circles
plt.grid() 
```

Task 2 involving energy levels. Somehow it wasnt entirely necessary to create a matrix equation, and seemed a lot less effort to do it without, using array.

```python
import numpy as np
import matplotlib.pyplot as plt

E1 = 1.0 #Energy values
E2 = 1.4
E3 = 1.2

w12 = abs(E2-E1) #omega values
w23 = abs(E3-E2)
w13 = abs(E3-E1)
h=1
t0=0 #start time
dt=0.1 #delta interval size
T=30 #end time

#Start conditions
N01=1
N02=0
N03=0

Nt = int(T/dt) #steps within interval up to T

#Arrays for N

N1=np.array([N01])
N2=np.array([N02])
N3=np.array([N03])

#Making a Time Array

t=np.array([t0])

#For loop and time development of Ns

for i in range(1,Nt+1):
    t=np.append(t,t[i-1]+dt)
Ntotal=np.array([N01+N02+N03])

for i in range(1,Nt+1):
    N1=np.append(N1,N1[i-1]+dt*(-w12*N1[i-1]+w13*N3[i-1]))
    N2=np.append(N2,N2[i-1]+dt*(-w23*N2[i-1]+w12*N1[i-1]))
    N3=np.append(N3,N3[i-1]+dt*(-w13*N3[i-1]+w23*N2[i-1]))

    Ntotal=np.append(Ntotal,N1[i]+N2[i]+N3[i])
plt.plot(t,N1)
plt.plot(t,N2)
plt.plot(t,N3)
plt.plot(t,Ntotal)
plt.legend(["N1","N2","N3","Ntotal"])
plt.show()
```
